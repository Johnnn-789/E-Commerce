• The Scenario: The company database crashed, and the backup data is a mess. You must clean the "Orders" dataset.
• Handling Missing Data: The dataset is riddled with NaNs. You must use .dropna() to remove rows completely missing 
an Order ID, .fillna() to replace missing shipping states and transaction with "Unknown" that were lost during export.

```python
import pandas as pd
df = pd.read_csv("ecommerce.csv")
df_dropany = df.dropna(how='any')
fill_shipping_state = df['shipping state'].fillna('unknown')
fill_trx = df['trx'].fillna('unknown')
print(fill_shipping_state)
print(fill_trx)
print(df_dropany)
```

• Handling Duplicates: A glitch caused users to be charged twice. You must use .duplicated() to find identical transactions and 
.drop_duplicates(keep='first') to refund/remove the errors.

```python
import pandas as pd
df = pd.read_csv("ecommerce.csv")
is_duplicated = df.duplicated()
df_dorp_first = df.drop_duplicates(keep='first')
print(df_dorp_first)
print(is_duplicated)
```

• Data Type Conversions: 
• "Price" is imported as a string (e.g., "$1,200.50"). You must remove the symbols and use pd.to_numeric() to turn it into a float.

```python
import pandas as pd
df = pd.read_csv("ecommerce.csv")
df['price'] = df['price'].str.replace('$', '', regex=False).str.replace(',', '', regex=False).astype(float)
df['price'] = pd.to_numeric(df['price'], errors='coerce')
print(df['price'])
```

• Replace the “pending” and “unknown” with the previous datetime

```python
df['purchase date'] = df['purchase date'].str.replace('pending', 'unknown')
print(df['purchase date'])
```

• Convert “purchase date" from messy text into a proper datetime object using pd.to_datetime().

```python
df['purchase date'] = pd.to_datetime(df['purchase date'], errors='coerce')
print(df['purchase date'])
```

• Convert the “purchase date" format from various format into this format “M/D/Y”

```python
df["purchase date"] = df["purchase date"].dt.strftime("%m/%d/%Y")
print(df['purchase date'])
```

```python
• String Manipulation: The "Names" column has chaotic capitalization and extra spaces. You use .str.title() 
and .str.strip() to fix this. You must also extract the email domain (e.g., extracting "gmail.com" from "user@gmail.com") using .str.split().
```

```python
import pandas as pd
df = pd.read_csv("ecommerce.csv")
df['customer name'] = df['customer name'].str.title().str.strip()
df['email_domain'] = df['email'].str.split("@").str[1]
print(df[['customer name', 'email', 'email_domain']].head())
```
